La cifra dominante de un entero es la primera por la izquierda que no es
nula. Encontrar los dı́gitos que pueden ser la cifra dominante al mismo
tiempo de 2 n y 5 n , con el mismo exponente n. Una vez encontrados, hay
que intentar demostrar que esos dı́gitos son los únicos posibles, pero, en
este caso, no me han sido de ninguna ayuda las pruebas que he hecho con
el ordenador.

```python
#haremos una funcion a la que le pasan un numero y te devuelve
#la cifra dominante
def func(x):
    a1=1
    flag1=0
    while(flag1==0):
        dom=0
        while(flag1==0 and  dom<10):
            dom=dom+1
            if(dom*a1>x):
                flag1=1
        a1=a1*10
    if(dom==1):
        return 9
    return dom-1
```

```python
L=[(n,func(2^n)) for n in srange(1,1000) if func(2^n)==func(5^n) ]
L
```
