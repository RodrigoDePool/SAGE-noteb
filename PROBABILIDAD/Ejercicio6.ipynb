```python
#Funcion que se encarga del intercambio de energia entre dos particulas aleatorias de la lista L
def intercambio(L):
        l = len(L);
        num1 = randint(0,l-1);
        num2 = randint(0,l-1);
        if(L[num1]<1): return L;
        L[num1] = L[num1] - 1;
        L[num2] = L[num2] + 1;
        return L;
```

```python
def interaccion(n,t):
    L = [5]*n;
    for i in range(0,t):
        L = intercambio(L);
    return L;
```
