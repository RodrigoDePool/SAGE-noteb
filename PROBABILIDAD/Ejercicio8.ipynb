```python
def movimiento():
    if(random() < 1/2):
        return -1;
    return 1;
def paseo(t):
    L=[0];
    ind = 0;
    for i in range(0,t-1):
        L.append(L[ind]+movimiento());
        ind = ind + 1;
    return L;
def media(L):
    l = len(L);
    suma = 0;
    for i in range(0,l):
        suma = suma + abs(L[i]);
    return suma/l;
```

```python
L = paseo(10000);
media = media(L);
```
