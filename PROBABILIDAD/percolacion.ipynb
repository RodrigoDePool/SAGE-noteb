```python
#Devuelve 1 con probabilidad p y 0 con probabilidad 1-p
def lanzamiento(p):
    if(random() < p):
        return 1;
    return 0;
```

```python
#Dado un punto p, devuelve su punto norte, sur , este u oeste del punto (en forma de lista)
def norte(p):
    L = [p[0], p[1]+1];
    return L;
def sur(p):
    L = [p[0], p[1]-1];
    return L;
def este(p):
    L = [p[0] + 1, p[1]];
    return L;
def oeste(p):
    L = [p[0]-1, p[1]];
    return L;
```

```python
#Con probabilidad prob anyade el punto 'punto' a la lista L
def intento_percolar(L,punto,prob):
    if(lanzamiento(prob) == 1):
        L.append(punto);
        return L;
    return L;
#Dado una lista con los elementos ya 'mojados', el punto que va a intentar esparcirse y la probabilidad con
#que eso ocurre, hacemos una iteracion de la percolacion sobre ese punto. Esto es agregar los puntos nuevos
#a los que percola el punto dado a la lista
def percolacion_iteracion(L,punto, prob):
    paux = norte ( punto );
    if(paux not in L):
        L = intento_percolar(L,paux,prob);
    paux = sur(punto);
    if(paux not in L):
        L = intento_percolar(L,paux,prob);
    paux = este(punto)
    if(paux not in L):
        L = intento_percolar(L,paux,prob);
    paux = oeste(punto);
    if(paux not in L):
        L = intento_percolar(L,paux,prob);
    return L;
```

```python
#Experimento de percolacion, termina cuando no puede percolar mas o cuando los puntos percoaldos son mas de n
def experimento_percolacion(prob, n):
    ind = 0;
    L = [[0,0]];
    while ind < len(L) and len(L) < n:
        p = L[ind];
        L = percolacion_iteracion(L,p,prob);
        ind = ind + 1;
    return L;
```
