Calcula probabilidad de obtener mas de 100 caras en 1000 lanzamientos donde la probabilidad de caras es de k/10 para k = 1, 2, 3 ... 9.

```python
#Cara == 1, Sello == 0. Sale cara con probabilidad p (argumento)
def lanzamiento(p):
    num = random();
    if(num < p):
        return 1;
    return 0;
```

```python
#Devuelve una lista con n lanzamientos de moneda, donde la probabilidad de cara es p
def lista_lanzamientos(n,p):
    L = [];
    for i in range(0,n):
        L.append(lanzamiento(p));
    return L;
```

```python
#Dada un k entero tal que 0 < k < 10, se realizan N tiradas de moneda con probabilidad de cara  k/10
#Si hay menos de l caras devuelve false, si hay mas devuelve true
def experimento(k,N, l):
    L = lista_lanzamientos(N,k/10);
    if(L.count(1) < l):
        return false;
    return true;
```

```python
#Dado k entero hace la simulacion con r repeticiones y devuelve la media del experimento propuesto en el ej1 1)
def simulacion_monedas(r,k):
        suma = 0;
        for i in range(0,r):
            if(experimento(k,1000,100) == true):
                suma = suma + 1;
        return suma/r;
```

```python
#Lanzamiento de moneda con probabilidad 1/2, si sale cara devuelve 1 y si sale cruz devuelve -1
def tirada():
    moneda = lanzamiento(1/2);
    if(moneda == 1):
        return 1;
    return -1;
```

```python
def juego(dinero):
    L=[100];
    jugador = dinero;
    while(jugador > 0):
        L.append(jugador);
        jugador = jugador + tirada();
    return L;
```

```python
def juego(dinero,tope, min):
    L=[100];
    jugador = dinero;
    while(jugador > min and jugador < tope):
        L.append(jugador);
        jugador = jugador + tirada();
    return (L,jugador);
```

```python
#r numero de repeticiones
def experimento2(r):
    suma = 0;
    for i in range(0,r):
        S = juego(100,200,50);
        if(S[1] > 100):
            suma = suma + 1;
    return suma/r;
```
