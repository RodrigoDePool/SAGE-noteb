```python
#Devuelve 1 si sube 0 si se mantiene -1 si baja
#Recibe n que indica si subio, bajo o se mantuvo en la ultima iteracion
#El n tiene la misma notacion que la devolucion
def cambios(n):
    if(n == 1):
        return cambioSubio();
    elif(n == -1):
        return cambioBajo();
    else:
        return cambioMantuvo();
    
def cambioSubio():
    n = random();
    if(n < 0.3):
        return 1;
    elif(n < 0.8):
        return -1;
    return 0;

def cambioBajo():
    n = random();
    if(n < 0.4):
        return 1;
    elif( n < 0.7):
        return -1;
    else:
        return 0;
    
def cambioMantuvo():
    n = random();
    if(n < 0.4):
        return 1;
    elif( n < 0.8):
        return -1;
    return 0;
```

```python
def simulacion_bolsa(t):
    bolsa = [0];
    ind = 0;
    subiendo = 0;
    bajando = 0;
    mantuvo = 0;
    for muda in range(0,t):
        cambio = cambios(bolsa[ind]) + bolsa[ind];
        ind = ind + 1;
        bolsa.append(cambio);
        if(cambio == 1):
            subiendo = subiendo + 1;
        elif(cambio == -1):
            bajando = bajando + 1;
        else:
            mantuvo =mantuvo +1;
    return (bolsa,subiendo,bajando,mantuvo);
```
