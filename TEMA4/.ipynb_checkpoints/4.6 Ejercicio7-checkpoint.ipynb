```python
#Definamos la funcion f
def f(x):
    L=[j^2 for j in x.digits()];
    return sum(L);
#Funcion que deveuelva la orbita de un numero por x  hasta que se cierre.
def orbita(x,f):
    L=[];
    aux=x;
    while not(aux in L):
        L.append(aux);
        aux=f(aux);
    return L;
```

```python
#Veamos las primeras 100 orbitas
for i in srange(1,10001):
    print  orbita(i,f);
```

Comprobamos rapidamente que todas las orbitas parecen cerrarse y decrecer rapidamente. Hemos visto que esto se cumple al menos para los primeros 10000 numeros.

Para probar que se cumple para todo numero de 6 digitos bastaria con ver que f(numero)<10000, como sabemos que el mayor numero que dara f para un numero de 6 digitos es f(999999)=486 ya sabemos que la orbita cerrara porque hemos probado para ese caso y  ,por tanto, cerrara para cualquier numero mas pequeno tambien. Ahora con induccion sobre el numero de digitos probemos que la orbita siempre cierra

Asumimos que la orbita cierra para numeros de n-1 digitos busquemos probar para numeros de n digitos. Bastaria con probar que 9^2+...+9^2 (n veces) se puede escribir con n-1 digitos (para n>=4 ya que para n=1,2,3 ya hemos probado nuestra conjetura) porque si es asi la demostracion acabaria por induccion. Veamos entonces que 89*n se puede escribir con n-1 digitos para n>=4

El caso base ya lo hemos probado (4).
Asumimos para numeros menores que n y probemos para n:
89n = 89*(n-1) + 89
Sabemos que el primer numero se puede escribir con n-2 digitos, esto seria como maximo el numero 99...999 con n-2 digitos si a este numero le sumamos 1 tenemos 1000...0 con n-1 digitos sabemos que al sumarle 88 seguira teniendo n-1 cifras ya que como poco n=4 y 100+88 sigue teniendo 3 cifras , si n es mayor esta claro que seguira teniendo n-1 cifras. Queda entonces demostrado.
